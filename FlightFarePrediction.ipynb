## clubbing these airlines together into category of 'other' since they have very low frequency

```python
Airline = train_df[['Airline']]
Current_Airline_List = Airline['Airline']
New_Airline_List = []

for carrier in Current_Airline_List:
  if carrier in ['Jet Airways','IndiGo','Air India','Multiple carriers','SpiceJet','Vistara','Air Asia','GoAir']:
    New_Airline_List.append(carrier)
  else:
    New_Airline_List.append('other')

Airline['Airline'] = pd.DataFrame(New_Airline_List)
Airline['Airline'].value_counts()
```

## Generating Dummy values for Airlines as we need numeric values to train our model

```python
Airline = pd.get_dummies(Airline, drop_first=True)
Airline = Airline.astype(int)
Airline.head()
```

```python
train_df['Destination'].value_counts()
```

## Clubbing the columns having destination as New Delhi and Delhi

```python
Destination = train_df[['Destination']]
Current_Destination_List = Destination['Destination']
New_Destination_List = []

for value in Current_Destination_List:
  if value in ['New Delhi']:
    New_Destination_List.append('Delhi')
  else:
    New_Destination_List.append(value)

Destination['Destination'] = pd.DataFrame(New_Destination_List)
```

## Calculation Variance Inflation Factor(VIF) for each feature

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(z):
  vif = pd.DataFrame()
  vif['variables'] = z.columns
  vif['VIF'] = [variance_inflation_factor(z.values, i) for i in range(z.shape[1])]
  return(vif)
```

```python
calc_vif(X)
```

## Year column has very high VIF and its feature importance is also 0 hence we remove it from the list of independent variables
